```python
from sklearn.cluster import KMeans
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import sklearn

from sklearn.datasets import load_iris
info = load_iris()
data = info.data
label = info.target
print(data.shape)
print(label.shape)
```

```python
def mypca(data, dim):
    std = sklearn.preprocessing.StandardScaler()
    data = std.fit_transform(data)
    m, n = np.shape(data)
    u, sigma, v = np.linalg.svd(data)
    return np.dot(data, v[:, :dim])
```

```python
data_2 = mypca(data, 2)

color = np.array(['yellow', 'lime', 'darkviolet'])

plt.scatter(data_2[:, 0], data_2[:, 1], c=color[label])
```
